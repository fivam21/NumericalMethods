```python
import numpy as np
import math
import pandas as pd
import scipy.stats as stats
import random
import matplotlib.pyplot as plt
```

```python
mu = 0.2
sigma = 0.1
nsample = 10000
x = np.linspace(0,2,101)

# Compute the PDF and CDF
#f = 1/(np.sqrt(2*np.pi)*sigma*x)*np.exp(-((np.log(x)-mu)/sigma)**2/2)
# F = 0.5*math.erfc((-1/np.sqrt(2)*(np.log(x)-mu)/sigma))
f = stats.cauchy.pdf(x,mu,sigma)
F = stats.cauchy.cdf(x,mu,sigma)
# plt.plot(x,f,'r', x,F,'b')
plt.plot(x,f,'r', label='PDF')
plt.plot(x,F,'b', label='CDF')
plt.legend(loc=0)
plt.title('Cauchy distribution with $\mu$ = 0.2 and $\sigma$ = 0.1')
# plt.savefig('pic.png')
```
