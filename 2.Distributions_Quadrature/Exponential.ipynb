```python
import numpy as np
import math
import pandas as pd
import scipy.stats as stats
import random
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mu = 2/3
nsample = 10000
x = np.linspace(0,5,nsample)

# f = 1/mu*np.exp(-x/mu)
# F = 1-np.exp(-x/mu)
# f = np.rand.exponential(mu,nsample)
# F = np.cumsum(f)/np.sum(f)
f = stats.expon.pdf(x, scale=mu)
F = stats.expon.cdf(x, scale=mu)
fig,axes = plt.subplots(figsize=(8,4))
axes.set_title('Exponential Distribution with $\mu$ = 2/3')
axes.plot(x,f, 'b', label='pdf')
axes.legend(loc=2)
axes2 = axes.twinx()
axes2.plot(x,F,'r', label='cdf')
axes2.legend(loc=1)
```

```python
# Sampling an exponential distribution
U = np.random.rand(nsample)
X = -mu*np.log(U)
plt.hist(X, bins=100, density=True, label='Sample')
plt.plot(x,f, 'r', label='Theory')
plt.legend()
plt.title('Sampling an exponential distribution with $\mu$ = 2/3')
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
# Plot a line chart for the sampled data instead of a histogram
U = np.random.rand(nsample)
x = np.linspace(0,5,6,1000)
f = stats.expon.pdf(x, scale=mu) #Theory
X = -mu*np.log(U) #Sample
plt.figure()
sns.kdeplot(X,label='Sample')
plt.plot(f, label='PDF')
plt.legend()
plt.title('Sampling an exponential distribution with $\mu$ = 2/3')
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
# Normalizing the histogram
dx = x[1]-x[0]
plt.hist((X/(nsample*dx)), bins=100, density=True)
```
