```python
import numpy as np
import math
import pandas as pd
import scipy.stats as stats
import random
import matplotlib.pyplot as plt
```

```python
# Sample distriubtion
mu, sigma = 0.2, 0.1
a,b = -0.4, 0.8 #truncation
ngrid=200
nsample=10^6
x=np.linspace(a,b,ngrid)

plt.figure(1,figsize=(3,3))
plt.plot(x,stats.norm.pdf(x,mu,sigma))
plt.xlim(a,b)
plt.title('PDF')

plt.figure(2,figsize=(3,3))
plt.plot(x,stats.norm.cdf(x,mu,sigma))
plt.xlim(a,b)
plt.title('CDF')
```

```python
nsample = 100000
U = np.random.rand(nsample)
# X = mu + sigma*stats.norm.ppf(U, 0, 1) # method 1a1: scale and shift
X = stats.norm.ppf(U, mu, sigma) # method 1a2
# X = mu + sigma*randn(nsample,1); % method 2
plt.hist(X, bins=50, density=True)
```

```python
# Empirical distribution
def norm_dist(x, mean, sd):
    prob_density = (np.pi*sd) * np.exp(-0.5*((x-mean)/sd)**2)
    return prob_density
    
x = np.linspace(a,b,ngrid)
f1 = norm_dist(x, mu, sigma)
plt.plot(x,f1)
```

```python
# Double distribution   
sample1 = np.random.normal(loc=20, scale=5, size=300)
sample2 = np.random.normal(loc=40, scale=5, size=700)
sample = np.hstack((sample1, sample2))
plt.hist(sample, bins=50)
```
